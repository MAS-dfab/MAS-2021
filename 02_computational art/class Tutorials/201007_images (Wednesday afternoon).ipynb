```python
import matplotlib.pyplot as plt
import math
import numpy as np
from PIL import Image
```

```python
img = Image.open('data/alpacas.jpg')
rotated_image = img.rotate(180)
img = rotated_image
pixel_values = img.load()
```

```python
#set size of figure
plt.figure(figsize=(20,20))

width = 800
height = 800

#set limits of axis
plt.xlim(0, 800)
plt.ylim(0, 800)

numx = 100
numy = 100

dx = float(width)/numx
dy = float(height)/numy

iw = img.width
ih = img.height

icx = float(iw)/numx
icy = float(ih)/numy

for y in range(numy):
    for x in range(numx):
        px = x*dx+dx/2
        py = y*dy+dy/2
        
        col = pixel_values[int(x*icx),int(y*icy)]
        
        my_color = [col[0]/256, col[1]/256, col[2]/256]
        cir = plt.Circle((px, py), dx/2, color = my_color)
        
        plt.gcf().gca().add_artist(cir)

```
