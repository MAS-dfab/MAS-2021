#Idea / Concept / Inspiration : Inspiration: The idea is to juxtapose the colors of a basic palette in a controlled manner that allows to observe their relationships with their surrounding and with each other.Inspiration: Sol Lewitt, Small Etchings

#Method : A line class describing a basic array defining the length of the segments, the distance between them and their number. The added functions help to control their position and display. Their color changes as the loop produces the geometry, giving us feedback on the procedure.
...

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import math
import random
import numpy as np
```

```python
class Line():
    
    def __init__(self, sx,sy, length, step, n):
        self.sx = sx
        self.sy = sy
        self.length = length
        self.step = step
        self.n = n
        
        self.x = []
        self.y=[]
    
    def drawGrid(self):
        
        for i in range (self.n):
            tempX = []
            tempY = []
            tempX.append(self.sx + i*self.step)
            tempX.append(self.sx + i*self.step)
            self.x.append(tempX)
            
            tempY.append(self.sy)
            tempY.append(self.sy + self.length)
            self.y.append(tempY)
            
                
    
    def rotate (self,angle):
        
        s = math.sin(angle)
        c = math.cos(angle)
        
        rotatedX = []
        rotatedY = []
        
        
        for listx,listy in zip(self.x,self.y):
            
            temp1 = []
            temp2 = []
            
            for x,y in zip(listx,listy):
                
                #move the point to origin
                x -= self.sx
                y -= self.sy
                #rotate
                xnew = x*c - y*s
                ynew = x*s + y*c
                
                #move it back to where it was
                x = xnew + self.sx
                y = ynew + self.sy
                
                temp1.append(x)
                temp2.append(y)
                
            rotatedX.append(temp1)
            rotatedY.append(temp2)
        
        
            
        self.x = rotatedX
        self.y = rotatedY

    
    
    def drawRed(self):
        
        for j, (x,y) in enumerate(zip(self.x, self.y)):
            r = j/self.n
            t = (j/self.n)*100
            plt.plot(x, y, color=(r,0,0, 0.5), linewidth=t/2)
            
    def drawGreen(self):
        
        for j, (x,y) in enumerate(zip(self.x, self.y)):
            g = j/self.n
            t = (j/self.n)*100
            plt.plot(x, y, color=(0,g,0, 0.5), linewidth=t/3)
            
    def drawBlue(self):
        
        for j, (x,y) in enumerate(zip(self.x, self.y)):
            b = j/self.n
            t = (j/self.n)*100
            plt.plot(x, y, color=(0,0,b, 0.5), linewidth=t/2)
            
    def drawYellow(self):
        
        for j, (x,y) in enumerate(zip(self.x, self.y)):
            r = g = j/self.n
            t = (j/self.n)*100
            plt.plot(x, y, color=(r,g*0.96,0,0.5), linewidth=t/3)
```

```python
fig = plt.figure(figsize = (100,100))
ax = fig.add_subplot(1, 1, 1) # nrows, ncols, index
cmap = mpl.cm.viridis
ax.set_facecolor( (cmap) )
# lime (0.78,1,0)
# grapefruit (0.96, 0.75, 0.8)
# nightsky (0.23, 0.26, 0.47)
# mint liquor (0.76, 0.78, 0.31)



l1 = Line(0,0,50, 2, 50)
l1.drawGrid()
l1.rotate(math.radians(20))
l1.drawBlue()

l2 = Line(0,0,50, 2, 50)
l2.drawGrid()
l2.rotate(math.radians(10))
l2.drawGreen()

l3 = Line(100,60,50, 2, 50)
l3.drawGrid()
l3.rotate(math.radians(-190))
l3.drawRed()

l4 = Line(90,60,50, 2, 50)
l4.drawGrid()
l4.rotate(math.radians(-180))
l4.drawYellow()
```
