```python
import matplotlib.pyplot as plt
import math
import random

import ipywidgets as widgets
from ipywidgets import interact
```

```python
##Class A: Polyline##

class Pline():
    
    def __init__(self,height,width,ltype,intensity):
        
        self.y1=height
        self.ltype=ltype
        self.intensity=intensity
        
        ys=[]
        xs=[]
        
        for i in range(0,width):
            xs.append(i*50)
            if self.ltype==1:
                ys.append(height+random.randint(-intensity,intensity))
            else:
                ys.append(height)
        
        self.x,self.y=xs,ys
    
    def display(self):
        plt.axis("equal")
        plt.plot(self.x,self.y, "black",linewidth = 1)
      
        
```

```python
##Class B: Pattern##

class Pattern():
    
    def __init__(self,pline1,pline2,density,number):
        
        self.pline1=pline1
        self.pline2=pline2
        self.density=density
        self.number=number
     
        self.used=[]
    
    def ortho(self):
        plt.axis("equal")
        cx=[]
        used1=[]
        
        while len(used1)<self.number:
            a= random.randint(0,len(self.pline1.x)-2)
            if not a in used1:
                used1.append(a)
                self.used.append(a)
        
        self.x=[]
        self.y=[]
        self.x2=[]
        self.y2=[]
     
        for i in used1:
            
            self.x.append(self.pline1.x[i])
            self.x2.append(self.pline2.x[i])
            self.y.append(self.pline1.y[i])
            self.y2.append(self.pline2.y[i])
            
            sum_dis= self.pline1.x[i+1]-self.pline1.x[i]
            step_dis= sum_dis/self.density
            
            
            sum_disy= self.pline1.y[i+1]-self.pline1.y[i]
            step_disy= sum_disy/(self.density+1)
            
            sum_disy2= self.pline2.y[i+1]-self.pline2.y[i]
            step_disy2= sum_disy2/(self.density+1)
            
            #random
            rang=[]
            
            for j in range(self.density):
                rang = random.randint(1,self.density-1)
                if rang%2==1:
                    cx=[self.pline1.x[i]+j*step_dis,self.pline2.x[i]+(j*step_dis)]
                    cy=[self.pline1.y[i]+j*step_disy,self.pline2.y[i]+j*step_disy2]
                    plt.plot(cx,cy, "black",linewidth = 0.5)

        
    def diagonal(self):
        
        cx=[]
        used1=[]
        
        while len(used1)<self.number:
            a= random.randint(0,len(self.pline1.x)-2)
            if not a in self.used:
                used1.append(a)
                self.used.append(a)
        
        self.x=[]
        self.y=[]
        self.x2=[]
        self.y2=[]
        self.used.extend(used1)
        for i in used1:
            
            self.x.append(self.pline1.x[i])
            self.x2.append(self.pline2.x[i])
            self.y.append(self.pline1.y[i])
            self.y2.append(self.pline2.y[i])
            
            sum_dis= self.pline1.x[i+1]-self.pline1.x[i]
            step_dis= sum_dis/self.density
            
            
            sum_disy= self.pline1.y[i+1]-self.pline1.y[i]
            step_disy= sum_disy/(self.density+1)
            
            sum_disy2= self.pline2.y[i+1]-self.pline2.y[i]
            step_disy2= sum_disy2/(self.density+1)
            
            allpx1=[]
            allpy1=[]
            
            allpx2=[]
            allpy2=[]
            
            for j in range(self.density):
                cx=[self.pline1.x[i]+j*step_dis,self.pline2.x[i]+(j*step_dis+random.randint(-10,10))]
                cy=[self.pline1.y[i]+j*step_disy,self.pline2.y[i]+j*step_disy2]
                
                allpx1.append(cx[0])
                allpy1.append(cy[0])
                
                allpx2.append(cx[1])
                allpy2.append(cy[1])
            
            for x,y in (zip(allpx1,allpy1)):
                for i in range(3):
                        r=random.randint(0,self.density-1)
                        plt.plot([x,allpx2[r]],[y,allpy2[r]], "black",linewidth = 0.5)
                        
                        
                        
                
                
                
```

```python
##Class C: Circle##
class Circle:
    def __init__(self,radius, x_pos, y_pos, color):
        self.radius = radius
        self.x_pos = x_pos
        self.y_pos = y_pos
        self.color = color
    def draw(self):
        res = 100
        a = 2* math.pi/res
        
        xc =[]
        yc = []
        for i in range(res):
            xc.append(self.x_pos + math.cos(i*a)*self.radius)
            yc.append(self.y_pos + math.sin(i*a)*self.radius)
        xc.append(xc[-1])
        yc.append(yc[-1])
        plt.plot(xc, yc, self.color,'po')
```

```python
##INTERACTIONS

@interact(seed = widgets.IntSlider(min=1, max=20, step=1, value=6),
         lines = widgets.IntSlider(min=1, max=20, step=1, value=12),
         intensity = widgets.IntSlider(min=1, max=50, step=1, value=15),
         pattern_number = widgets.IntSlider(min=1, max=10, step=1, value=2),
         pattern_density = widgets.IntSlider(min=1, max=20, step=1, value=5))

def show(seed,lines,intensity,pattern_number,pattern_density):
    
    #figureSize
    plt.figure ( figsize =(5,5))
    #set the limits of axis
    plt.xlim(-50,250)
    plt.ylim(-50,250)
    
    #POLYLINES#########################################################
    
    plines=[]
    height=0
    
    for i in range(lines-1):

        if i!=0 and i!=lines-2:
            p=Pline(height,lines,1,intensity)
        else:
            p=Pline(height,lines,0,0)
        plines.append(p)
        height=plines[-1].y1+lines*5+ random.randint(-10,3)
    for p in plines:
        p.display()

    #PATTERNS###########################################################

    #ortho
    pats=[]
    for i in range(len(plines)-1):
        pat=Pattern(plines[i],plines[i+1],pattern_density,pattern_number)
        pats.append(pat)
    for pa in pats:
        pa.ortho()

    #diagonal 
    pats2=[]
    for i in range(len(plines)-1):
        pat=Pattern(plines[i],plines[i+1],pattern_density,pattern_number)
        pats2.append(pat)
    for pa in pats2:
        pa.diagonal()
        
    #SHAPES#################################3
    #rectangle

    rx=[]
    ry=[]

    for i,j in (zip(plines[-1].x,plines[-1].y)):
                xmax=max(plines[-1].x)
                ymax=max(plines[-1].y)
                rx,ry=[0,xmax,xmax,0,0],[0,0,ymax,ymax,0]
    #Circles
    c=[]
    c= plines[1].x

    for i in range(seed):
        ci=Circle(random.randint(5,40),c[random.randint(1,len(c)-2)],random.randint(30,ymax),"black")
        ci.draw()

    

    plt.plot(rx, ry, "black",'po',linewidth = 2)            
```
