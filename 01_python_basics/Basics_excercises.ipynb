```python
import math
import matplotlib.pyplot as plt
```

# Loops

```python
########################
###### grid

plt.figure(figsize = (15,15))
plt.axis('equal')

N = 25
x,y = [],[]
d = 10.0

for i in range(N):
    for j in range(N):
        x.append(i*d)
        y.append(j*d)

plt.plot(x, y, 'ro')
```

```python
########################
###### diagonal

plt.figure(figsize = (15,15))
plt.axis('equal')

N=12
x,y = [],[]
d = 1.0

for i in range(N):
    x.append(i*d)
    y.append(i*d)
    
plt.plot(x,y,'ro', color='green')

for i in range(N):
    x[i] += 1.0
plt.plot(x,y,'ro', color='red')

for i in range(N):
    y[i] += 2.0
plt.plot(x,y,'ro', color='blue')
```

```python
########################
###### square

plt.figure(figsize = (15,15))
plt.axis('equal')

Nx=4
Ny=6
d = 1.0

x1, y1 = [],[]
x2, y2 = [],[]
x3, y3 = [],[]
x4, y4 = [],[]

# plt.yscale(10)
# 
for i in range(Nx):
    x1.append(i*d)
    y1.append(0)
plt.plot(x1,y1,'ro', color='red', markersize=10) 

for j in range(Ny):
    x2.append((Nx-1)*d)
    y2.append(j*d)
plt.plot(x2,y2,'ro', color='blue', markersize=10) 


for i in range(Nx):
    x3.append(i*d)
    y3.append((Ny-1)*d)
plt.plot(x3,y3,'ro', color='green', markersize=10) 

for j in range(Ny):
    x4.append(0)
    y4.append(j*d)
plt.plot(x4,y4,'ro', color='black', markersize=10) 

## Question: what happened to the corners? Why? 
```

```python
########################
###### circle

plt.figure(figsize = (15,15))
plt.axis('equal')

N = 40
x,y = [],[]
r = 10.0

for i in range(N):
    theta = math.pi * 2 * i/N
    x.append(r * math.cos(theta))
    y.append(r * math.sin(theta))
    
plt.plot(x,y,'ro', color='red', markersize=10) 

## Question: if we draw the circle as a line, why is it open?
## How can we close it? 
```
