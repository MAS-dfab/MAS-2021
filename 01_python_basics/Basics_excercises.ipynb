```python
import math
import matplotlib.pyplot as plt
```

# Loops

```python
plt.figure(figsize = (15,15))
plt.axis('equal')

N = 25
x,y = [],[]
d = 10.0

for i in range(N):
    for j in range(N):
        x.append(i*d)
        y.append(j*d)

plt.plot(x, y, 'ro')

```

```python
plt.figure(figsize = (15,15))
plt.axis('equal')

N=12
x,y = [],[]
d = 1.0

for i in range(N):
    x.append(i*d)
    y.append(i*d)
    
plt.plot(x,y,'ro', color='green')

for i in range(N):
    x[i] += 1.0
plt.plot(x,y,'ro', color='red')

for i in range(N):
    y[i] += 1.0
plt.plot(x,y,'ro', color='blue')
```

```python
plt.figure(figsize = (15,15))
plt.axis('equal')

Nx=4
Ny=6
d = 1.0

x1, y1 = [],[]
x2, y2 = [],[]
x3, y3 = [],[]
x4, y4 = [],[]

plt.yscale(10)

for i in range(Nx):
    x1.append(i*d)
    y1.append(0)
plt.plot(x1,y1,'ro', color='red', markersize=10) 

for j in range(Ny):
    x2.append((Nx-1)*d)
    y2.append(j*d)
plt.plot(x2,y2,'ro', color='blue', markersize=10) 


for i in range(Nx):
    x3.append(i*d)
    y3.append((Ny-1)*d)
plt.plot(x3,y3,'ro', color='green', markersize=10) 

for j in range(Ny):
    x4.append(0)
    y4.append(j*d)
plt.plot(x4,y4,'ro', color='black', markersize=10) 

```
